```python
top = json.load(open('../save3/top_atv.json', 'r'))
df = pd.DataFrame(columns=['year', 'type', 'name', 'h_index'])
name_l, h_l, year_l, type_l = [], [], [], []
names = ['author', 'topic', 'venue']
for year in range(2000, 2022):
    temp = top[str(year)]
    
    for name in names:
        author = temp[name]
        author_name = [a[0] for a in author]
        author_h = [a[1] for a in author]
        name_l.extend(author_name)
        h_l.extend(author_h)
        year_l.extend([year]*len(author_h))
        type_l.extend([name]*len(author_h))
df['name'] = name_l
df['h_index'] = h_l
df['year'] = year_l
df['type'] = type_l
print(df.describe())
```

```python
df.sort_values(by=['year','type']).to_csv('../save3/top_atv.csv', index=None)
```
